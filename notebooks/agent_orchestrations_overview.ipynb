# Quorum Across Orchestrations

Map candidate outputs into Quorum and use the same consensus API across orchestration layers.

```python
from quorum_core import AgentOutput, resolve_consensus

candidates = [
    AgentOutput(id="orchestrator-1", content="draft A", confidence=0.3),
    AgentOutput(id="orchestrator-2", content="draft A", confidence=0.6),
    AgentOutput(id="orchestrator-3", content="draft B", confidence=0.9),
]

resolve_consensus(candidates, mode="quantum_ready")
```

## Integration Rule

If a system can emit a list of candidates, it can use Quorum. That includes CrewAI, LangGraph, AutoGen, OpenClaw, and custom agent graphs.
