# AutoGen Integration

Use Quorum after a multi-agent discussion loop or any orchestration layer that emits final candidates.

```python
from quorum_core import AgentOutput, resolve_consensus

candidates = [
    AgentOutput(id="agent-a", content="yes", confidence=0.4),
    AgentOutput(id="agent-b", content="yes", confidence=0.8),
    AgentOutput(id="agent-c", content="no", confidence=0.9),
]

resolve_consensus(candidates, mode="weighted_majority")
```

## Final Answer Gate

A common pattern is to let the conversation happen first, then use Quorum as the final gate before answering the user.
