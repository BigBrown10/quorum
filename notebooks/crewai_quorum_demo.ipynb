# CrewAI Integration

Connect Quorum to CrewAI and other orchestration systems that emit candidate outputs.

```python
from quorum_core import AgentOutput, resolve_consensus

candidates = [
    AgentOutput(id="crew-1", content="42", confidence=0.7),
    AgentOutput(id="crew-2", content="42", confidence=0.9),
    AgentOutput(id="crew-3", content="7", confidence=0.2),
]

result = resolve_consensus(candidates, mode="quantum_ready")
result
```

## Adapter Pattern

Map each agent answer into `AgentOutput`, call Quorum after the crew finishes, and inspect `unstable` before returning a final answer to the user.

```python
def crewai_adapter(agent_results):
    candidates = [
        AgentOutput(id=item['id'], content=item['content'], confidence=item.get('confidence'))
        for item in agent_results
    ]
    return resolve_consensus(candidates, mode="quantum_ready")
```
