# LangGraph Integration

Use Quorum as a reducer or post-processing step for LangGraph or any graph-based orchestration layer.

```python
from quorum_core import AgentOutput, resolve_consensus

graph_outputs = [
    AgentOutput(id="node-1", content={"answer": "red"}, confidence=0.3),
    AgentOutput(id="node-2", content={"answer": "red"}, confidence=0.6),
    AgentOutput(id="node-3", content={"answer": "blue"}, confidence=0.8),
]

resolve_consensus(graph_outputs, mode="quantum_ready")
```

## Graph Reduction

Use Quorum as the final reducer after multiple branches have produced candidate answers. The same reducer pattern works for other systems that emit a list of candidates.
